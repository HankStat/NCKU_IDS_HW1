```python
import pandas as pd
df = pd.read_csv('IMDB-Movie-Data.csv')
first = df[df.Year ==2016].sort_values(by='Rating',ascending = False)
first.head(3)
#取所有2016的電影 並依據rating做排序
```

```python
#print(df.info())
#print(df.loc[0]['Actors'])
ator = []
for i in range(len(df['Actors'])):
    actorlist = df.loc[i]['Actors'].split("|")
    for j in range(len(actorlist)):
        dict = {}
        actorlist[j] = actorlist[j].strip()
        dict['actor'] = actorlist[j]
        dict['revenue'] = df.loc[i]['Revenue (Millions)']
        ator.append(dict)
act_rev = pd.DataFrame(ator)
act_rev2 = act_rev.fillna(0.0)
group = act_rev2.groupby(act_rev2.actor)
second = group.aggregate(pd.DataFrame.mean).sort_values(by='revenue',ascending = False)
#second
second.head(2)
#因為actors裡面很多actor 所以要取出來並split 
#在把缺失值補0 在以actor進行分群 去計算每個actor 的平均收入並做排序
```

```python
#print(df.info())
ew=[]
for i in range(len(df['Actors'])):
    actorlist = df.loc[i]['Actors'].split("|")
    for act in actorlist:
        act = act.strip()
        if act == "Emma Watson":
            ew.append(df.loc[i]['Rating'])
third = pd.Series(ew)
print(third.mean())
#把有艾瑪華生的資料取出來 並把對應的rating收集起來最後取平均
```

```python
dtoa = []
for i in range(len(df['Actors'])):
    actorlist = df.loc[i]['Actors'].split("|")
    for act in actorlist:
        act = act.strip()
        empty = {}
        empty['director'] = df.loc[i]['Director']
        empty['actor'] = act
        dtoa.append(empty)
fourth = pd.DataFrame(dtoa)
fourth.drop_duplicates(keep='last',inplace = True)
fourth.groupby("director").count().actor.sort_values(ascending = False).head(4)
#fourth是一個dataframe裡面每筆資料都是合作的director 跟 actor
#把重複actor的刪掉並依據actor的數量做排序
```

```python
atog = []
for i in range(len(df['Actors'])):
    actorlist = df.loc[i]['Actors'].split("|")
    genrelist = df.loc[i]['Genre'].split("|")
    for act in actorlist:
        act = act.strip()
        for genre in genrelist:
            empty = {}
            empty['actor'] = act
            empty['genre'] = genre
            atog.append(empty)
fifth = pd.DataFrame(atog)
fifth.drop_duplicates(keep='last',inplace = True)
fifth.groupby("actor").count().genre.sort_values(ascending = False).head(6)
#fifth是一個dataframe裡面每筆資料都是合作的actor 跟 genre
#把重複genre的刪掉並依據genre的數量做排序
```

```python
gap = []
for i in range(len(df['Actors'])):
    actorlist = df.loc[i]['Actors'].split("|")
    for act in actorlist:
        act = act.strip()
        empty = {}
        empty['actor'] = act
        empty['year'] = df.loc[i]['Year']
        gap.append(empty)
sixth = pd.DataFrame(gap)
grouped = sixth.groupby(sixth.actor)
sixthmax = grouped.aggregate(pd.DataFrame.max)
sixthmin = grouped.aggregate(pd.DataFrame.min)
Sixth = pd.merge(sixthmax, sixthmin, on='actor')
#先創造兩個seires 分別為演員所演過的電影的最近年分 跟最遠年分
#print(Sixth.info())
Sixth['GAP'] = Sixth ['year_x'] - Sixth['year_y']
Sixth.sort_values(by='GAP',ascending = False).head(53)
#創造一個column並把最近年分-最遠年分 並依據gap做排序
```

```python
re = []
for i in range(len(df['Actors'])):
    actorlist = df.loc[i]['Actors'].split("|")
    for z in range(len(actorlist)):
        actorlist[z] = actorlist[z].strip()
    for j in range(len(actorlist)):
        for k in range(len(actorlist)):
            if j!=k:
                empty = {}
                empty['actor'] = actorlist[j]
                empty['correspond'] = actorlist[k]
                re.append(empty)
seven = pd.DataFrame(re)
seven.drop_duplicates(keep='last',inplace = True)
k = seven[seven.actor == "Johnny Depp"]
for i in range(len(k)):
    temp = seven[seven.actor == k.iloc[i]['correspond']]
    for j in range(len(temp)):
        temp.iloc[j]['actor'] = 'Johnny Depp'
    result = pd.concat([k,temp])
    k = result
k
```
